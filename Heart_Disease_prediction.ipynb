Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and procressing

```python
heart_data = pd.read_csv('/heart.csv')
```

```python
heart_data.head()
```

Checking number of rows and columns

```python
heart_data.shape
```

Getting Info of Data

```python
heart_data.info()
```

Statistical measure of data

```python
heart_data.describe()
```

Checking distrubution of target value

```python
heart_data['HeartDisease'].value_counts()
```

1---defective heart
0---healthy heart

So spliting feature and targets


```python
X = heart_data.drop(columns='HeartDisease',axis=1)
Y = heart_data['HeartDisease']
```

Spliting traning and testing Data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Now model traning

Logistic regression

```python
model = LogisticRegression()
```

```python
print(X_train.dtypes)
```

# Identify non-numeric columns and apply one-hot encoding
# This will convert categorical columns into numerical columns


```python
X_train = pd.get_dummies(X_train, drop_first=True) # drop_first=True avoids multicollinearity
X_test = pd.get_dummies(X_test, drop_first=True)
```

```python
model.fit(X_train,Y_train)
```

Evaluation matrics

Accuracy score of traning data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data :',training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data :',test_data_accuracy)
```

BUlding predictive system

```python
input_data = (63,1,3,145,233,1,0,150,0,2.3,0,0,1)

# Convert the tuple to a pandas DataFrame with the ORIGINAL column names from heart_data
# Based on heart_data.head() and info(), the original column names are:
# 'Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS', 'RestingECG', 'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope', 'Ca', 'Thal'
# We need to map the input tuple values to these names.
# Let's assume the tuple maps directly to the order of columns after dropping 'HeartDisease'
# from heart_data in the cell where X and Y were created.
# Printing heart_data.columns before dropping 'HeartDisease' confirms the column order:
# Index(['Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS',
#        'RestingECG', 'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope', 'Ca',
#        'Thal', 'HeartDisease'], dtype='object')
# So the columns for X are:
input_columns_for_X = ['Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS', 'RestingECG', 'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope', 'Ca', 'Thal']

input_df = pd.DataFrame([input_data], columns=input_columns_for_X)

# Apply the same one-hot encoding as used for the training data (X_train).
# Based on the original data and common heart datasets, the categorical columns likely are:
# 'Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope', 'Ca', 'Thal'
# (Note: 'Ca' and 'Thal' were likely treated as categorical in the original dataset if they represent specific categories)
# Let's use the columns identified as categorical in the original thought process, adjusting for original casing.
# We need to determine which columns in the original heart_data were actually treated as categorical.
# Looking at the dtypes of heart_data would confirm this. Assuming the user's prior
# assumption was correct about which columns were categorical, but correcting for casing:
original_categorical_cols = ['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope', 'Ca', 'Thal'] # Adjusted based on assumed original categorical columns

# Apply get_dummies to the input DataFrame
# Use the original column names for 'columns' parameter
input_data_processed = pd.get_dummies(input_df, columns=original_categorical_cols, drop_first=True)

# Align the columns of the processed input data with the columns of the training data (X_train)
# This is crucial to ensure the model receives features in the expected order and format.
# Use reindex to add missing columns (filled with 0) and drop extra columns.
input_data_processed = input_data_processed.reindex(columns=X_train.columns, fill_value=0)


# Now make the prediction using the processed input data
prediction = model.predict(input_data_processed)

print(prediction)
if(prediction[0]==0):
  print('The person does not have heart disease')
else:
  print('The person has heart disease')
```
